# Conditional Statement 

```python
if condition: 
    # do something 
```

```python
# if statement (one condition)
num = 10 
if num > 0:
     print("positive")
```

```python
# if-else statement (two condition)
num = -10 
if num > 0:
     print("positive")
else:
    print("negative")
```

```python
# if..elif..else statement (more than two conditions)
num = 0
if num > 0:
     print("positive")
elif num == 0: 
    print("zero")
else:
    print("negative")
```

```python
num = int(input("Enter your number..."))
if num > 0:
     print("positive")
else:
    print("negative")
```
